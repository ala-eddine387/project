```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "http://books.toscrape.com"
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.text,"html.parser")
titles =[link["title"] for link in soup.find_all("a",title = True, limit = 25)]
prices = [price.text[1:] for price in soup.find_all("p",class_ = "price_color", limit = 25)]
availables =[avaib.text.strip() for avaib in soup.find_all("p",class_ = "instock availability", limit = 25)]


```

* in the the prices list i use slice the first caractere in the string because it has no relation to the price (&)
* in the availables list i use split() method because the list contain a lot of space tab and newline

```python
df = pd.DataFrame({
    "Title" : titles,
    "Price": prices,
    "Availability" : availables 
})
df.to_csv("books.csv",index = False)
```
